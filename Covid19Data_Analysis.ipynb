# Exploratory Data Analysis

### State 

```python
df.columns
```

```python
cov_statc = df.groupby(['State'], as_index= False)['Confirmed'].sum().sort_values(by = 'Confirmed', ascending= False).head(10)
sns.set(rc={'figure.figsize':(20,5)})

sns.barplot(x= 'State', y= 'Confirmed', data = cov_statc)
```

***From above graph we can say that most confirmed cases were recorded from states like New York,New Jersy, Hubei respectively***

```python
cov_statd = df.groupby(['State'], as_index= False)['Deaths'].sum().sort_values(by = 'Deaths', ascending= False).head(10)
sns.set(rc={'figure.figsize':(20,5)})

sns.barplot(x= 'State', y= 'Deaths', data = cov_statd)
```

***From above graph we can say that most deaths were recorded from states like New York,New Jersy, Hubei respectively***

```python
cov_statr = df.groupby(['State'], as_index= False)['Recovered'].sum().sort_values(by = 'Recovered', ascending= False).head(10)
sns.set(rc={'figure.figsize':(20,5)})

sns.barplot(x= 'State', y= 'Recovered', data = cov_statr)
```

***From above graph we can say that most patients recovered in states like New York,Hubei, New South Wales respectively***

###  Region 

```python
df.columns
```

```python
cov_regc = df.groupby(['Region'], as_index= False)['Confirmed'].sum().sort_values(by = 'Confirmed', ascending= False).head(5)
sns.set(rc={'figure.figsize':(10,5)})

sns.barplot(x= 'Region', y= 'Confirmed', data = cov_regc)
```

***From above graph we can say that most confirmed cases were recorded from regions like US,Mainland China, Canada respectively***

```python
cov_regd = df.groupby(['Region'], as_index= False)['Deaths'].sum().sort_values(by = 'Deaths', ascending= False).head(5)
sns.set(rc={'figure.figsize':(10,5)})

sns.barplot(x= 'Region', y= 'Deaths', data = cov_regd)
```

***From above graph we can say that most deaths were recorded from regions like US,Mainland China, Canada respectively***

```python
cov_regr = df.groupby(['Region'], as_index= False)['Recovered'].sum().sort_values(by = 'Recovered', ascending= False).head(5)
sns.set(rc={'figure.figsize':(10,5)})

sns.barplot(x= 'Region', y= 'Recovered', data = cov_regr)
```

***From above graph we can say that most patients recovered in states like US,Mainland China, Canada, Australia respectively***

# Conclusion

***We conclude that the confirmed cases, deaths on 29th April 2020 were recorded mostly in US, Mainland China, Canada regions and recoveries were mostly recorded in the previously mentioned regions including Australia...***
